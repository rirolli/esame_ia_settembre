```python
import pandas as pd
import sklearn as sl
import numpy as np

from sklearn.model_selection import train_test_split
```

## Caricamento del dataset completo

Caricamento del dataset heart_failure_clinical_records_dataset.train.csv.

```python
url = 'heart_failure_clinical_records_dataset.train.csv'

app_data = pd.read_csv(url, na_values=np.nan)
```

```python
app_data.head(15)
```

## Split del dataset

Dividiamo il dataset in due parti: Train e Test dataset.

```python
# Dividiamo il dataset in due parti: Train e Test
train, test = train_test_split(app_data, test_size=.2)
```

Salviamo i dataset appena creati in file CSV.

```python
train.to_csv('train.csv', index=False)
test.to_csv('test.csv', index=False)
```
